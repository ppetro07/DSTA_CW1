```python
# Correlation coefficient is a useful measure to suggest which features should be given more attention in further investigation.
#Another method of identifying the important features is through the use of Random Forrest Classifier
```

```python
# Selection of important features using Random Forrest Classifier
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
train, test = train_test_split(df, test_size = 0.3)
print(train.shape)
print(test.shape)
```

```python
#All available prediction features are included as this is an investigation to select 
#the ones which would be most useful for prediction
pred_features= ['diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']
```

```python
train_X = train[pred_features]
train_y = train.diagnosis
test_X= test[pred_features]
test_y =test.diagnosis
```

```python
#Fitting of Random Forest Classifier
model=RandomForestClassifier(n_estimators=100)
model.fit(train_X,train_y)
```

```python
#Imporance of each feature for prediction
featimp = pd.Series(model.feature_importances_, index=pred_features).sort_values(ascending=False)
print(featimp)
```

```python
# Graphical representing the most important features for prediction.
plt.figure(figsize=(16,16))
plt.title("Feature importances (Random Forests)")
plt.bar(pred_features, featimp,)

plt.xticks(rotation='vertical')
plt.subplots_adjust(bottom=0.3)


```

```python
# Using the Random Forrest Classifier method the "mean" features are indicated as important
#Based on the plot above the important features indicated are: radius_mean, texture_mean, perimeter_mean, area_mean, smoothness_mean,
#compactness_mean,concavity_mean, concave points_mean
```

```python
# Further investigation of distribution of the fields suggested as important from the Random Forest Classifier output
features_mean=list(df)
# split dataframe into two based on diagnosis
dfM=df[df['diagnosis'] ==1]
dfB=df[df['diagnosis'] ==0]
```

```python
# Histogram distribution of features for both Malignant and Benign classes. 
plt.rcParams.update({'font.size': 8})
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(10,10))
axes = axes.ravel()
for idx,ax in enumerate(axes):
    ax.figure
    binwidth= (max(df[features_mean[idx]]) - min(df[features_mean[idx]]))/50
    ax.hist([dfM[features_mean[idx]],dfB[features_mean[idx]]], bins=np.arange(min(df[features_mean[idx]]), max(df[features_mean[idx]]) + binwidth, binwidth) , alpha=0.5,stacked=True, normed = True, label=['M','B'])
    ax.legend(loc='upper right')
    ax.set_title(features_mean[idx])
plt.tight_layout()
plt.show()
```

```python
#The distinguishing power of a feature for the purpose of classification can be inferred by how distinguishing a feature is
#for the two classification sets. Hence, distinguishing features are the ones with differing means for Malignant and Benign groups
# Mean values for radius, perimeter, area, compactness, concavity and concave points show best separation between target classes
#This is consistent with the information obtained from Random Forrest Classifier and Correlation Matrix, as these fields
#are included in the output of all selection methods.
```
